```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Mumbai.csv')
df.head()
print(df.shape)
```

```python
df2 = df.copy()
df2.columns
```

```python
df2.describe()
```

```python
sns.boxplot(df2['Area'])
plt.show()
```

```python
sns.boxplot(df2['No. of Bedrooms'])
```

**Handling the Outliers**

```python
sns.distplot(df2.Price,bins=100)
```

```python
features = ['Area','No. of Bedrooms']
for feature in features:
    percentile75 = df2[feature].quantile(0.75)
    percentile25 = df2[feature].quantile(0.25)
    IQR = percentile75 - percentile25
    upper_limit = percentile75 + (1.5*IQR)
    lower_limit = percentile25 - (1.5*IQR)
    df2.loc[df2[feature]<=lower_limit,feature] = lower_limit # capping the outliers
    df2.loc[df2[feature]>=upper_limit,feature] = upper_limit

```

```python
sns.boxplot(df2.Area)
```

 Removed all the outliers from Area column

```python
sns.boxplot(df2['No. of Bedrooms'])
```

Removed all the outliers from No. of Bedrooms column

```python
df2 = df2.drop_duplicates()
df2.shape
```
