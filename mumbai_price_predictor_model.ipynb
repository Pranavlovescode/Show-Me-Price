**Analysing the values of the Location label**

```python
filtered_counts = df['Location'].value_counts()[df['Location'].value_counts() < 10]
print(filtered_counts)
```

**Converting the values of the Location label which are less than 10 to 'other'**

```python
df['Location'] = df['Location'].apply(lambda x: 'Other' if df['Location'].value_counts()[x] <= 10 else x)
df
```
