# **Mumbai Price Detector ML model**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mumbai1.csv')
df.head()
```

## **Exploratory Data Analysis**

```python
df.isnull().sum()
```

```python
print(df['Location'].value_counts())
print(df.shape)
```

```python
filtered_counts = df['Location'].value_counts()[df['Location'].value_counts() < 50]
print(filtered_counts)
filtered_counts = df['Location'].value_counts()[df['Location'].value_counts() > 50]
print(filtered_counts)
```

```python
df['Location'] = df['Location'].apply(lambda x: 'Other' if df['Location'].value_counts()[x] <= 50 else x)
df
```

```python
df2 = df.drop(['Unnamed: 0'],axis=1)
print(df2.shape)
```

**One Hot Encoding to convert categorical data into numbers**

```python
from sklearn.model_selection import train_test_split
X = df2.iloc[:,1:18]
y =df2.iloc[:,0:1]
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
y_train
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse_output=False)
```

```python
X_train_new = ohe.fit_transform(X_train[['Location']])
X_train_new
# X_train_new.shape
```

```python
X_test_new = ohe.fit_transform(X_test[['Location']])
X_test_new
# X_test_new.shape
```

```python
X_train_final = np.concatenate((X_train_new,X_train),axis=1)
X_train_final
```

```python
X_test_final = np.concatenate((X_test_new,X_test),axis=1)
X_test_final
```

**Training the model**

```python
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(sparse_output=False,drop='first'),['Location'])
],remainder='passthrough')
```

```python
X_train_transformed = transformer.fit_transform(X_train)
```

```python
X_test_transformed=transformer.fit_transform(X_test)
```

**Predicting using Linear Regression**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train_transformed,y_train)
```

```python
lr.score(X_test_transformed,y_test)
```

```python
lr.predict(X_test_transformed)
```

```python
y_test
```

**Predicting the model using SVM**

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train_transformed,y_train)
```

```python
svm.score(X_test_transformed,y_test)
```

**Predicting using Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
r = RandomForestClassifier()
r.fit(X_train_transformed,y_train)
```

```python
r.score(X_test_transformed,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
logr.fit(X_train_transformed,y_train)
```

```python
logr.score(X_test_transformed,y_test)
```
